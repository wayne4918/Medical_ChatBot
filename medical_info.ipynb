```python
def doctor_bayes(adj_mat, symptom_list, symptoms, diseases):
    sym = [symptoms.index(s) for s in symptom_list]
    p_dis = adj_mat.sum(axis=0) / adj_mat.sum()
    p_sym = adj_mat.sum(axis=1) / adj_mat.sum()
    dist = []
    for i in range(len(diseases)):
        prob = np.prod((adj_mat[:, i] / adj_mat[:, i].sum())[sym]) * p_dis[i] / np.prod(p_sym[sym])
        dist.append(prob)
    if sum(dist) == 0:
        return 'I have no idea'
    else:
        idx = dist.index(max(dist))
        return diseases[idx]

# 인접 행렬 생성
adj_mat = np.zeros((len(symptoms), len(diseases)))

for i in range(len(data)):
    disease = data.iloc[i, 0]
    for j in range(len(symptom_columns)):
        symptom = data.iloc[i, j+1]
        if pd.notnull(symptom):
            sym_idx = symptoms.index(symptom)
            dis_idx = diseases.index(disease)
            adj_mat[sym_idx, dis_idx] += 1

# 예시 실행
print(doctor_bayes(adj_mat, ['오한','떨림'], symptoms, diseases))
```

```python
import pandas as pd
import numpy as np
import torch
from transformers import AutoModelForSequenceClassification, AutoTokenizer, TrainingArguments, Trainer
from sklearn.model_selection import train_test_split



# 증상 열 추출
symptom_columns = [col for col in data.columns if '증상' in col]

# 고유한 증상 추출
unique_symptoms = pd.Series(data[symptom_columns].values.ravel()).dropna().unique()

# 질병 리스트
diseases = list(data['질병'].unique())

# 증상 리스트 생성
symptoms = list(unique_symptoms)

# 데이터프레임 생성
mat = np.zeros((len(data), len(symptoms) + 1))

for i in range(len(data)):
    mat[i, 0] = diseases.index(data.iloc[i, 0])
    for j in range(len(symptom_columns)):
        symptom = data.iloc[i, j+1]
        if pd.notnull(symptom):
            sym_idx = symptoms.index(symptom)
            mat[i, sym_idx + 1] = 1  # 증상이 있으면 1로 표시

symptoms_li = ['질병'] + symptoms
di_sym_df = pd.DataFrame(mat, columns=symptoms_li)
train_df, test_df = train_test_split(di_sym_df, test_size=0.1, stratify=di_sym_df['질병'], random_state=2022)

class DiseaseDataset(torch.utils.data.Dataset):
    def __init__(self, df):
        self.labels = df.iloc[:, 0].values
        self.texts = df.iloc[:, 1:].apply(lambda row: ' '.join([symptom for symptom, present in zip(symptoms, row) if present == 1]), axis=1).values
        self.encodings = tokenizer(list(self.texts), truncation=True, padding=True, max_length=128)

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx]).long()
        return item

# 모델과 토크나이저 로드
model_name = "bert-base-multilingual-cased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=len(diseases))

train_dataset = DiseaseDataset(train_df)
test_dataset = DiseaseDataset(test_df)

# 트레이너 설정
training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=3,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    warmup_steps=500,
    weight_decay=0.01,
    logging_dir='./logs',
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
)

# 모델 훈련
trainer.train()
```
